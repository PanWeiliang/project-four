```python
df.head()
```

```python
df.salary=salary
df.tail()
```

```python
c=CurrencyConverter()
```

```python
index=df[df.country=='United Kingdom'].salary.index
salary=df[df.country=='United Kingdom'].salary.apply(lambda x: c.convert(x,'GBP', 'USD'))
```

```python
df.ix[index,'salary']=salary
```

```python
index=df[df.country=='Australia'].salary.index
salary=df[df.country=='Australia'].salary.apply(lambda x: c.convert(x,'AUD', 'USD'))
df.ix[index,'salary']=salary
```

```python
df.tail()
```

```python
index=df[df.country=='Canada'].salary.index
salary=df[df.country=='Canada'].salary.apply(lambda x: c.convert(x,'CAD', 'USD'))
df.ix[index,'salary']=salary
```

```python
salary_median = np.median(df.salary)
salary_median
```

```python
# plot histogram of salary distribution
df.salary.plot(kind='hist', bins=range(10000, 300001, 10000), 
               title='Distribution of Salary',
               fontsize=11, color='blue',figsize=(20,18))
plt.xlabel('Annual Salary ($)')
plt.xlim(0,300000)
plt.xticks(range(0, 300001, 50000), ['0k', '50k', '100k', '150k', '200k', '250k', '300k'])
# add median and mean lines
plt.axvline(np.median(df.salary), linestyle='dashed', c='yellow', label='median')
plt.axvline(np.mean(df.salary), linestyle='dotted', c='orange', label='mean')
plt.legend(loc='best')
```

### Salary above median salary will be classified as high salary

```python
df['high_salary'] = df.salary > salary_median
```

## QUESTION 1: Factors that impact salary
## Using country as feature to predict salary

```python
X=pd.get_dummies(df.country).drop('United Kingdom',axis=1)
```

```python
y=df.high_salary
y.value_counts()
```

```python
clif= RandomForestClassifier(n_estimators=1000, random_state=0, n_jobs=-1)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
```

```python
scores = cross_val_score(clif, X_train, y_train, cv=5, scoring='accuracy')
print('Accuracy score for logistic regression is {:.4f} +/- {:.4f}'.format(np.mean(scores), np.std(np.std(scores))))
```

```python
# Train the classifier
clif.fit(X_train, y_train)
clif.score(X_test,y_test)
```

```python
y_pred=clif.predict(X_test)
print(pd.crosstab(pd.Series(y_test,name="Actual"), pd.Series(y_pred,name='Predicted'), margins=True))
print(classification_report(y_test, y_pred))
```

```python
logreg = LogisticRegression()
scores = cross_val_score(logreg, X_train, y_train, cv=10, scoring='accuracy')
print('Accuracy score for logistic regression is {:.4f} +/- {:.4f}'.format(np.mean(scores), np.std(np.std(scores))))
```

```python
logreg.fit(X_train,y_train)
```

```python
y_pred1=logreg.predict(X_test)
print(pd.crosstab(pd.Series(y_test,name="Actual"), pd.Series(y_pred1,name='Predicted'), margins=True))
print(classification_report(y_test, y_pred1))
```

## Using job description as feature to predict salary

```python
cv = CountVectorizer(ngram_range=(1,2), max_features=2500, binary=True, stop_words='english')
description = cv.fit_transform(df.description)
```

```python
cv.get_feature_names()
```

```python
desc_df = pd.DataFrame(description.toarray(), columns=cv.get_feature_names())
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(desc_df, y, train_size=0.7)
```

```python
# Train the classifier
clif.fit(X_train1, y_train1)
clif.score(X_test1,y_test1)
```

```python
y_pred2=clif.predict(X_test1)
print(pd.crosstab(pd.Series(y_test1,name="Actual"), pd.Series(y_pred2,name='Predicted'), margins=True))
print(classification_report(y_test1, y_pred2))
```

```python
logreg = LogisticRegression()
scores = cross_val_score(logreg, X_train1, y_train1, cv=10, scoring='accuracy')
print('Accuracy score for logistic regression is {:.4f} +/- {:.4f}'.format(np.mean(scores), np.std(np.std(scores))))
```

```python
logreg.fit(X_train1,y_train1)
y_pred3=logreg.predict(X_test1)
print(pd.crosstab(pd.Series(y_test1,name="Actual"), pd.Series(y_pred3,name='Predicted'), margins=True))
print(classification_report(y_test1, y_pred3))
```

```python
fi_clif = pd.DataFrame({
        'feature':desc_df.columns,
        'importance':clif.feature_importances_
    })

fi_clif.sort_values('importance', ascending=False, inplace=True)
```

```python
fi_clif.head(20)
```

## Key words in job descriptions that influence pay

```python
fi_clif.head(20).plot(kind='bar', y='importance',x='feature', rot=90, 
                             figsize=(18,13), title='Feature importances', 
                             sort_columns=False)
```

### Job description is a better feature to predict salary

```python
logreg.fit(X_train_res,y_train_res)
y_pred3=logreg.predict(X_test)
print(pd.crosstab(pd.Series(y_test,name="Actual"), pd.Series(y_pred3,name='Predicted'), margins=True))
print(classification_report(y_test, y_pred3))
```

```python
dtrain=xgboost.DMatrix(X_train,label=y_train)
dtest=xgboost.DMatrix(X_test)
train_labels=dtrain.get_label()
ratio=float(np.sum(train_labels==False))/np.sum(train_labels==True)
params={'max_depth':1,'silent':1,'eta':1,'scale_pos_weight':ratio}
```

```python
bst=xgboost.train(params,dtrain,)
y_pred4=(bst.predict(dtest)>0.5)
```

```python
print(pd.crosstab(pd.Series(y_test,name="Actual"), pd.Series(y_pred4,name='Predicted'), margins=True))
print(classification_report(y_test, y_pred4))
```
